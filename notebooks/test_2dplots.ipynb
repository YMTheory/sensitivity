# 2D plotting in python
I am using `pcolormesh` method in python matplotlib to plot the two-dimensional $\Delta \chi^2$ plot, while the coordinate orders of x-axis and y-axis requires to be confirmed here.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mat = np.zeros((4, 4))
for i in range(4):
    for j in range(4):
        mat[i, j] = i*2 + j
print(mat)
```

```python
fig, ax = plt.subplots(figsize=(7,5 ))
x_arr = range(4)
y_arr = np.arange(4) + 5
im = ax.pcolormesh(y_arr, x_arr, mat, )
cb = plt.colorbar(im, ax=ax)
```

Here are the conclusions I got from the test:

- The inputs of axes in `pcolormesh` (the first two arguments) should be the X-axis coordinates (left to right) and Y-axis coordinates (bottom to top);

- The input of the value matrix (the third argument) should be the matrix itself. It will be reverted along the axis 0 so that the 1st row we assigned to the matrix goes to the bottom (not on the top as we print the matrix directly). Therefore, we do not need to convert the matrix (along axis 0) by ourselves.
