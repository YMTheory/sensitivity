```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('/Users/yumiao/Documents/Works/0nbb/nEXO/Physics/sterile_nu/sensitivity/')

from xsection import xsection
```

```python
xsec = xsection()
xsec.load_csv()
```

```python

plt.style.use('ggplot')

Enu_1 = np.arange(0.70, 5, 0.01)
xsec_1 = xsec.interp_xsec_state1(Enu_1)
Enu_2 = np.arange(1.0, 5, 0.01)
xsec_2 = xsec.interp_xsec_state2(Enu_2)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(Enu_1, xsec_1, lw=3, label='1+, 590 keV')
ax.plot(Enu_2, xsec_2, lw=3, label='1+, 850 keV')
ax.semilogy()
ax.set_xlabel('Neutrino Energy [MeV]')
ax.set_ylabel(r'$\sigma_{E_x} \, [cm^2]$')
ax.legend(fontsize=13)
```
